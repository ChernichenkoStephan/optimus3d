## Задача 1. 

```python
f1 = function('2*(x1^2) + x1*x2 + 3*(x2^2)')
f1
```

```python
x_0 = np.array([-1, 1])
a = 0.01
eps = 0.5
a = eps = 0.5

stop_criteria = union_stopper(
    iterations_stopper(100) , 
    grad_norm_stopper(eps),
)

optimizer = GradientOptimizer(
    stopper=stop_criteria,
    stepper=armijo_stepper(a, eps),
    verbose=True,
)
x_min, f1_min, trajectory = optimizer.min(f1, x_0)
```

## Задача 4. 

```python
f4 = function('x1^2 - 4*x1 + x2^2 - 2*x2')
f4
```

```python
intervals = ((0, 1), (0, 2))
eps = 0.1

stop_criteria = union_stopper(
    iterations_stopper(50),
    grad_norm_stopper(eps),
    step_norm_stopper(eps),
)

optimizer = CGradientOptimizer(
    stopper=stop_criteria,
    stepper=iteration_stepper(),
    verbose=True,
)
x_min, f4_min, trajectory = optimizer.min(f4, intervals)
```

```python
f4_min
```

```python
ax = surface_figure(intervals=intervals, slide=-4)
```

```python
f4.show(
    ax=ax,
    point=f4_min,
    radius=1.2,
    trajectory=trajectory,
)
```

### Дополнительно (Шаг с методом Armijo)

## Задача 5. 

```python
f5 = function('2*x1^2 + (x2 - 1)^2')
f5
```

```python
f_condition = function('2*x1 + x2')
f_condition
```

```python
a = 0.5
eps = 0.1

stop_criteria = union_stopper(
    iterations_stopper(50),
    step_norm_stopper(eps),
)

optimizer = QuadraticPenaltyOptimizer(
    stopper=stop_criteria,
    c_stepper=iteration_stepper(lambda k: 2**k),
    a_stepper=const_stepper(1/16),
    verbose=True,
)
x_min, f5_min, trajectory = optimizer.min(
    f5,
    f_condition, 
    x_0=np.array([1, 1]),
)
```

```python
f5_min
```

```python
raduis = 1.5
ax = f_condition.figure(
    radius=raduis,
)
```

```python
f5.show(
    ax=ax,
    point=f5_min,
    radius=raduis,
    trajectory=trajectory,
)
```
